```python
import matplotlib.pyplot as plt
import numpy as np
import time

T1 = time.time()
# Function to compute the mandelbrot set
def Mandelbrot( x , y , MAX ):
    
    # Use of complex function to set y to be a complex number
    C = complex(x,y)
    
    # 1jy is the same as iy
    Z = x + 1j*y 
    
    for i in range( MAX ):
        Z = Z*Z + C
        if(abs(Z)>2):
            return i
  
    return MAX

# Variables set for both x and y, to create a desired resolution in the x-y plane
X_res = int(input("Enter a value for your desired resolution:   "))
Y_res = X_res 

# Variable received through input for maximum number of iterations
MAX = int(input("Enter how many iterations you would like to run:   "))

# Fills a 2-dimensional array of a size in which is dictated by the resolution
list_XY = np.zeros([ X_res , Y_res ])

for list_X , x in enumerate(np.linspace( -2 , 0.7 , X_res)):
    for list_Y , y in enumerate(np.linspace( -1.2 , 1.2 , Y_res)):
        list_XY[ list_X , list_Y ] = Mandelbrot( x , y , MAX)
            
plt.figure( dpi = 175 )
plt.imshow( list_XY.T , cmap = 'CMRmap' , interpolation = 'bilinear' , extent = [ -2 , 0.7 , -1.2 , 1.2 ] )
plt.xlabel('Reals')
plt.ylabel('Complex')
plt.show()
T2 = time.time()
print("{0:.3}".format(T2-T1) , " Seconds")
```
